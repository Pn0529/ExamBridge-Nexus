```python
import pdfplumber

# Replace this with your actual file name
file_name = "syllabus.pdf"

with pdfplumber.open(file_name) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        print(text)
```

```python
import re

def clean_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)

    return text
```

```python
cleaned_text = ""

with pdfplumber.open(file_name) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            cleaned_text += clean_text(text)

print(cleaned_text[:1000])  # print first 1000 characters
```

```python

gate_text = """
network theory ohms law kirchhoffs laws impedance power factor
renewable energy wind turbines earthing rms sinusoidal waveform
electrical circuits analysis
"""
```

```python
cleaned_gate_text = clean_text(gate_text)
print(cleaned_gate_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

documents = [cleaned_text, cleaned_gate_text]

tfidf_matrix = vectorizer.fit_transform(documents)

print("TF-IDF Shape:", tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print("Similarity Percentage:", similarity[0][0] * 100)
```

```python
gate_topics = [
    "network theory",
    "ohms law",
    "kirchhoffs laws",
    "impedance",
    "power factor",
    "renewable energy",
    "wind turbines",
    "earthing methods",
    "rms value",
    "electrical circuits"
]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

results = []

for topic in gate_topics:

    documents = [cleaned_text, topic]

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(documents)

    similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])

    results.append((topic, similarity[0][0]))

# Print results
for topic, score in results:
    print(topic, "->", score)
```

```python
# Convert to percentage
results_percent = [(topic, score * 100) for topic, score in results]

# Sort by highest similarity
results_percent.sort(key=lambda x: x[1], reverse=True)

for topic, score in results_percent:
    print(f"{topic} -> {score:.2f}%")
```

```python
# Sort by LOWEST similarity (high priority)
results_percent.sort(key=lambda x: x[1])

print("🔥 High Priority Topics (Low Coverage):\n")

for topic, score in results_percent:
    print(f"{topic} -> {score:.2f}%")
```

```python
# Get lowest 3 similarity topics
weak_topics = results_percent[:3]

print("🔥 Recommended Topics To Study:\n")

for topic, score in weak_topics:
    print(f"{topic} ({score:.2f}%)")
```

```python
print("\n📺 Recommended YouTube Searches:\n")

for topic, score in weak_topics:
    search_query = topic.replace(" ", "+")
    youtube_link = f"https://www.youtube.com/results?search_query=GATE+{search_query}+full+lecture"
    print(f"{topic} -> {youtube_link}")
```

```python
def get_top_video(topic):
    search_request = youtube.search().list(
        q=f"GATE {topic} full lecture",
        part="snippet",
        type="video",
        maxResults=5
    )
    search_response = search_request.execute()

    video_ids = [item['id']['videoId'] for item in search_response['items']]

    video_request = youtube.videos().list(
        part="statistics,snippet",
        id=",".join(video_ids)
    )
    video_response = video_request.execute()

    videos = []

    for item in video_response['items']:
        title = item['snippet']['title']
        stats = item['statistics']

        views = int(stats.get('viewCount', 0))
        likes = int(stats.get('likeCount', 0))
        video_id = item['id']
        link = f"https://www.youtube.com/watch?v={video_id}"

        # Smart ranking score
        score = views + (likes * 10)

        videos.append((title, views, likes, score, link))

    videos.sort(key=lambda x: x[3], reverse=True)

    return videos[0]
```
